```python
## importing project modules
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# load data
df = pd.read_csv("all_data.csv")
print(df.head())
df.shape
```

```python
print(df.Country.unique())
```

```python
print(df.Year.unique())
```

```python
# renaming column for coding purposes
df = df.rename({"Life expectancy at birth (years)":"LEABY"}, axis = "columns")
df.head()
```

```python
# plotting data by GDP
plt.figure(figsize=(8,6))
sns.histplot(df.GDP)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
# plotting data by Life expectancy
plt.figure(figsize=(8,6))
sns.histplot(df.LEABY)
plt.xlabel("Life expectancy at birth (years)")
```

```python
dfMeans = df.drop("Year", axis = 1).groupby("Country").mean().reset_index()
dfMeans
```

```python
# Life Expectancy per Country Bar Plot
plt.figure(figsize=(8,6))
sns.barplot(x="LEABY", y="Country", data=dfMeans)
plt.xlabel("Life expectancy at birth (years)")
```

```python
# GDP per Country Bar Plot
plt.figure(figsize=(8,6))
sns.barplot(x="GDP", y="Country", data=dfMeans)
plt.xlabel("GDP in Trillions of U.S. Dollars")
```

```python
# Violin Plots for GDP and Life expectancy
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.violinplot(ax=axes[0], x=df.GDP, y=df.Country, hue=df.Country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[1] = sns.violinplot(ax=axes[1], x=df.LEABY, y=df.Country, hue=df.Country)
axes[1].set_xlabel("Life expectancy at birth (years)")
```

```python
# Swarm Plots for GDP and Life expectancy
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.swarmplot(ax=axes[0], x=df.GDP, y=df.Country, hue=df.Country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[1] = sns.swarmplot(ax=axes[1], x=df.LEABY, y=df.Country, hue=df.Country)
axes[1].set_xlabel("Life expectancy at birth (years)")
```

```python
# Overlaying violin plots and swarm plots
fig, axes = plt.subplots(1, 2, sharey=True, figsize=(15, 5))
axes[0] = sns.violinplot(ax=axes[0], x=df.GDP, y=df.Country,color = "black", hue=df.Country)
axes[0] = sns.swarmplot(ax=axes[0], x=df.GDP, y=df.Country, hue=df.Country)
axes[0].set_xlabel("GDP in Trillions of U.S. Dollars")
axes[1] = sns.violinplot(ax=axes[1], x=df.LEABY, y=df.Country, color = "black", hue=df.Country)
axes[1] = sns.swarmplot(ax=axes[1], x=df.LEABY, y=df.Country, hue=df.Country)
axes[1].set_xlabel("Life expectancy at birth (years)")
```

```python
# Line plots for GDP
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.GDP, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars")
```

```python
# Separated line plots for GDP
graphGDP = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False)

graphGDP = (graphGDP.map(sns.lineplot,"Year","GDP")
         .add_legend()
         .set_axis_labels("Year","GDP in Trillions of U.S. Dollars"))

graphGDP
```

```python
# Line plots for Life Expectancy
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.LEABY, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("Life expectancy at birth (years)")
```

```python
# Separated line plots for Life Expectancy
graphLEABY = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False)

graphLEABY = (graphLEABY.map(sns.lineplot,"Year","LEABY")
         .add_legend()
         .set_axis_labels("Year","Life expectancy at birth (years)"))

graphLEABY
```

```python
# Scatter plot of life expectancy and GDP
sns.scatterplot(x=df.LEABY, y=df.GDP, hue=df.Country).legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
```

```python
# Separated scatter plot of life expectancy and GDP
graph = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False, sharex = False)
graph = (graph.map(sns.scatterplot,"LEABY", "GDP")
         .add_legend()
         .set_axis_labels("Life expectancy at birth (years)", "GDP in Trillions of U.S. Dollars"))
```
